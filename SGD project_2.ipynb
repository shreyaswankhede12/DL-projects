```python
import torch
import numpy as np
import matplotlib.pyplot as plt
```

```python
D1=torch.normal(0.0, 1.0,size=[500,2])
D2=torch.normal(7.2, 1.0,size=[500,2])
```

```python
plt.scatter(D1[:,0], D1[:,1])
plt.scatter(D2[:,0], D2[:,1])
```

```python
X=torch.empty(1000,D1.size(1)+1).fill_(1)
X[::2,:2] = D1
X[1::2,:2] = D2

Y=torch.empty(1000,1)
Y[::2,:2] = 1.0
Y[1::2,:2] = 0.0
print (X.size(0))
```

```python
W = torch.zeros(X.size(1))
```

```python
x = np.linspace(-4,8,500)
y = np.zeros(500,) 
plt.plot(x, y, '-r', label='W')
plt.title('Graph of W before training')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.scatter(D1[:,0], D1[:,1])
plt.scatter(D2[:,0], D2[:,1])
plt.grid()
plt.show()
```

```python

def sigmoid(z):
          return 1/(1+np.exp(-z))
            
```

```python
def gradient_descent(x,y,w,eta,max_epochs):
    for i in range(max_epochs):
        g=0
        
        for j in range(X.size(0)):
            g=g+ (sigmoid(x[j].dot(w))-y[j])*x[j]
        g=g/1000
        w=w-eta*g
    return w
        
        
```

```python
W=gradient_descent(X,Y,W,0.1,5000)
print(W)
```

```python
x = np.linspace(-4,10,1000)
y = -(W[0].item()/W[1].item())*x - (W[2].item()/W[1].item())
plt.plot(x, y, '-r', label='W')
plt.title('Graph of W after training')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.scatter(D1[:,0], D1[:,1])
plt.scatter(D2[:,0], D2[:,1])
plt.grid()
plt.show()
```
